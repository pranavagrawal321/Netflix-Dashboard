```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
df["listed_in"] = df["listed_in"].str.split(",")
```

```python
df.head()
```

```python
value_counts = df['listed_in'].apply(len).value_counts()

if len(value_counts) > 1:
    df_exploded = df.explode("listed_in")
else:
    df_exploded = df.copy()
```
